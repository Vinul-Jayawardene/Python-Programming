```python
import tkinter as tk 

window = tk.Tk()
window.title("My GUI")
window.geometry("800x600")

label = tk.Label(window, text="Hello Tkinter")
label.pack(pady=10)

button = tk.Button(window, text="Click me")
button.pack(pady=10)

entry = tk.Entry(window)
entry.pack(pady=10)

text = tk.Text(window)
text.pack(pady=10)


window.mainloop()


```

```python
import tkinter as tk 

window = tk.Tk()
window.title("My GUI")
window.geometry("800x600")


entry = tk.Entry(window)
entry.grid(row=0, column=1, padx=10, pady=5)

entry = tk.Entry(window)
entry.grid(row=0, column=2, padx=10, pady=5)




window.mainloop()
```

```python
import tkinter as tk 

window = tk.Tk()
window.title("My GUI")
window.geometry("800x600")


button = tk.Button(window, text="Click me")
button.place(y=100, x=500)

window.mainloop()
```

```python
import tkinter as tk 

window = tk.Tk()
window.title("My GUI")
window.geometry("800x600")

def say_hello():
    print("Hello")

button = tk.Button(window, text="Click me", command=say_hello)
button.pack()

window.mainloop()
```
